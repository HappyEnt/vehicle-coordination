```python
# import numpy,matplotlib etc.
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

names_map = {
    0: 'PRE_REGULARIZATION',
    1: 'POST_REGULARIZATION',
    2: 'PROGRESSIVE_PRE_REGULARIZATION',
    3: 'PROGRESSIVE_POST_REGULARIZATION',
    4: 'NON_PARAMETRIC_BP',
    5: 'RESAMPLE_MOVE',
    6: 'SIR',
    7: 'SIR_ROUGHENING',
}
```

```python
%matplotlib widget

data = np.loadtxt('particles-vs-accuracy-post.csv', delimiter=',', skiprows=1)

# extract first column to particles, second column to filter_type and last column to error
particles = data[:,0]
filter_type = data[:,1]
error = data[:,2]
duration_millis = data[:,3]

error_dict = {}
millis_dict = {}
for i in range(len(particles)):
    if (particles[i], filter_type[i]) in error_dict:
        error_dict[(particles[i], filter_type[i])].append(error[i])
        millis_dict[(particles[i], filter_type[i])].append(duration_millis[i])
    else:
        error_dict[(particles[i], filter_type[i])] = [error[i]]
        millis_dict[(particles[i], filter_type[i])] = [duration_millis[i]]

error_list = list(error_dict.values())

millis_list = list(millis_dict.values())
plt.rcdefaults()
# plt.xkcd()
        
config_names = list(error_dict.keys())
# map ids to names using names_map
#config_names = [(config[0], names_map[config[1]]) for config in config_names]
config_names = [config[0] for config in config_names]
plt.style.use('bmh')
colors = ['lightblue', 'lightblue', 'lightblue', 'lightblue', 'lightblue', 'lightblue', 'lightblue']
boxplot = plt.boxplot(error_list,
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     )
plt.title('Post-Regularization: Accuracy vs Particles')
for patch, color in zip(boxplot['boxes'], colors):
    patch.set_facecolor(color)

# add config_names to each boxplot
plt.xticks(range(1, len(config_names)+1), config_names)
# add title
# set x label to 'Particles'
plt.xlabel('Particles')
# set y label to "Error [m]"
plt.ylabel('Error [m]')
```
